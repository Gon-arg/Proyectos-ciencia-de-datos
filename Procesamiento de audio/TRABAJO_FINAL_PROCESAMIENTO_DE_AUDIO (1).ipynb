```python
# Instalar librerías necesarias (solo una vez)
#!pip install gTTS SpeechRecognition pydub transformers torch --quiet

from gtts import gTTS
from pydub import AudioSegment
import speech_recognition as sr
from transformers import pipeline
from IPython.display import Audio, display
import os

# Paso 1: Convertir texto a voz (.mp3)
file_path = "/content/texto1.txt"
output_mp3_path = "texto1.mp3"
output_wav_path = "texto1.wav"

try:
    with open(file_path, "r", encoding="utf-8") as f:
        text_to_speak = f.read()
except FileNotFoundError:
    print(f"Error: El archivo {file_path} no fue encontrado.")
    text_to_speak = None

if text_to_speak:
    tts = gTTS(text=text_to_speak, lang='es', slow=False)
    tts.save(output_mp3_path)
    print(f"Audio guardado como '{output_mp3_path}'")

    # Convertir MP3 a WAV PCM para transcripción
    sonido = AudioSegment.from_mp3(output_mp3_path)
    sonido = sonido.set_frame_rate(16000).set_channels(1).set_sample_width(2)
    sonido.export(output_wav_path, format="wav")
    print(f"Audio convertido a WAV como '{output_wav_path}'")

    # Reproducir audio original
    display(Audio(output_mp3_path))

# Paso 2: Transcripción del audio
def transcribir_audio(audio_path):
    recognizer = sr.Recognizer()
    with sr.AudioFile(audio_path) as source:
        audio_data = recognizer.record(source)
        try:
            texto = recognizer.recognize_google(audio_data, language="es-ES")
            print("Transcripción:")
            print(texto)
            return texto
        except sr.UnknownValueError:
            print("No se pudo entender el audio.")
        except sr.RequestError as e:
            print(f"Error al conectar con el servicio de reconocimiento: {e}")
    return None

# Paso 3: Análisis de sentimiento
def analizar_sentimiento(texto):
    analizador = pipeline("sentiment-analysis", model="pysentimiento/robertuito-sentiment-analysis")
    resultado = analizador(texto)[0]
    sentimiento = resultado["label"]
    confianza = round(resultado["score"] * 100, 2)
    print(f"\nSentimiento detectado: {sentimiento} ({confianza}%)")
    return sentimiento, confianza

# Paso 4: Generar resumen y convertirlo a audio
def generar_audio_resumen(sentimiento, confianza, salida_audio):
    resumen = f"El sentimiento principal del audio es {sentimiento.lower()} con una confianza del {confianza} por ciento."
    tts = gTTS(text=resumen, lang='es', slow=False)
    tts.save("temp_resumen.mp3")

    # Convertir a WAV PCM
    sonido = AudioSegment.from_mp3("temp_resumen.mp3")
    sonido = sonido.set_frame_rate(16000).set_channels(1).set_sample_width(2)
    sonido.export(salida_audio, format="wav")
    print(f"\n Resumen de sentimiento guardado como '{salida_audio}'")
    return salida_audio

# Paso 5: Ejecutar pipeline completo
texto_transcrito = transcribir_audio(output_wav_path)
if texto_transcrito:
    sentimiento, confianza = analizar_sentimiento(texto_transcrito)
    audio_resumen = generar_audio_resumen(sentimiento, confianza, "resumen_sentimiento.wav")
    display(Audio("resumen_sentimiento.wav"))
else:
    print(" No se pudo completar el pipeline por falta de transcripción.")
```

```python
from gtts import gTTS
from pydub import AudioSegment
from IPython.display import Audio, display
import datetime

# Menú de la pizzería
menu = {
    "1": "Pizza Margarita",
    "2": "Pizza Napolitana",
    "3": "Pizza Cuatro Quesos",
    "4": "Pizza Fugazzeta",
    "5": "Pizza Calabresa"
}

# Generar texto del menú
texto_menu = "Bienvenido a la pizzería. Nuestro menú es el siguiente: "
for k, v in menu.items():
    texto_menu += f"Opción {k}: {v}. "
texto_menu += "Por favor, escriba el número de la opción que desea ordenar."

# Crear audio
tts = gTTS(text=texto_menu, lang='es')
tts.save("menu.mp3")

# Convertir a WAV PCM
sonido = AudioSegment.from_mp3("menu.mp3")
sonido = sonido.set_frame_rate(16000).set_channels(1).set_sample_width(2)
sonido.export("menu.wav", format="wav")

# Reproducir
display(Audio("menu.wav"))
print("Escuchá el menú y luego ejecutá el siguiente bloque para responder.")

def generar_audio(texto, nombre_archivo):
    tts = gTTS(text=texto, lang='es')
    tts.save(nombre_archivo + ".mp3")

    sonido = AudioSegment.from_mp3(nombre_archivo + ".mp3")
    sonido = sonido.set_frame_rate(16000).set_channels(1).set_sample_width(2)
    sonido.export(nombre_archivo + ".wav", format="wav")

    return Audio(nombre_archivo + ".wav")

def guardar_pedido(nombre, direccion, pedido_texto):
    hora = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    with open("pedido_cliente.txt", "w", encoding="utf-8") as f:
        f.write(f"Nombre: {nombre}\n")
        f.write(f"Dirección: {direccion}\n")
        f.write(f"Pedido: {pedido_texto}\n")
        f.write(f"Hora del pedido: {hora}\n")
    print("Pedido guardado en 'pedido_cliente.txt'")
```

```python
pedido_escrito = input("Escribí el número de la opción que querés ordenar: ")
pedido_final = menu.get(pedido_escrito.strip(), "Opción inválida")
print(f"Pedido seleccionado: {pedido_final}")
```

```python
display(generar_audio("¿Cuál es su nombre?", "pregunta_nombre"))
```

```python
nombre_cliente = input("Escribí tu nombre: ")
```

```python
display(generar_audio("¿Cuál es su dirección de entrega?", "pregunta_direccion"))
```

```python
direccion_cliente = input("Escribí tu dirección de entrega: ")
```

```python
resumen = f"Gracias {nombre_cliente}. Usted ha pedido {pedido_final} para entregar en {direccion_cliente}. Su pedido llegará en aproximadamente 30 minutos."
display(generar_audio(resumen, "resumen_final"))
guardar_pedido(nombre_cliente, direccion_cliente, pedido_final)
```

elegimos hacer un chatbot para un local de pizeria, en este caso adaptado para google colab con codigo segmentado por bloques, y realizamos transcripciones de texto a audio  utilizando la libreria SpeechRecognition. De igual forma se puede generar un chatbot para una clinica, un club de depotes, un local de ropas, una rotiseria, o cualquier empresa u organizacion que desea automatizar una parte de sus flujos de trabajo. Las limitaciones que tiene son que que te ofrece opciones ya establecidas y comparado con la atencion de una persona es las respuestas espontaneas que puede dar la misma. El chatbot solo responde entre algunas opciones, esta limitado por su programmacion, aunque muchas veces suene similar a las personas no deja de ser un algoritmo programado. Como contraparte uno de los beneficios es que el chatbot no se enoja ni se alegra puede soportar tanto a clientes conformes como a clientes no conformes o enojados, evitando el estres que se genera en las personas que realizan atencion al cliente. Por lo expuesto vemos que el chatbot tiene limitaciones pero tambien tiene beneficios, quedara a criterio de cada organizacion si desea implementarlo o no.
Como mejoras podemos integrar al chatbot una caficacion del cliente posterior a la entrega, para ver si le gusto la pizza o llego en condiciones, de esta forma se evaluaria la calidad del proceso de entrega y la calidad del proceso de elaboracion, para poder ver si hay fallas y mejorar la parte del proceso mas criticada o con baja calificacion.
